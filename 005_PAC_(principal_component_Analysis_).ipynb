```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import pandas as pd
iris = load_iris()
X = iris.data
y = iris.target
X.shape
pd.DataFrame(X)
pd.DataFrame(y)

pca = PCA(n_components=2)#实例化
pca = pca.fit(X) #拟合模型
X_dr = pca.transform(X)
print(X_dr)
# X_dr = PCA(2).fit_transform(X)
 
plt.figure()
plt.scatter(X_dr[y == 0, 0], X_dr[y == 0, 1], c="blue", label=iris.target_names[0],alpha=.7 )
plt.scatter(X_dr[y == 0, 1], X_dr[y == 1, 1], c="red", label=iris.target_names[1] ,alpha=.7)
plt.scatter(X_dr[y == 2, 0], X_dr[y == 2, 1], c="green", label=iris.target_names[2],alpha=.7 )
plt.xlabel("X_dr[x1]")
plt.ylabel("X_dr[x2]")
plt.legend()
plt.title("PCA OF IRIS DATASET")
plt.show()
print(pca.explained_variance_)
print(pca.explained_variance_ratio_)# ratio of every new eigen variance / total variance
print(pca.explained_variance_ratio_.sum())#eigen variance / total variance

```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
iris = load_iris()
X = iris.data
y = iris.target
X.shape
pd.DataFrame(X)
pd.DataFrame(y)

# pca_f = PCA(n_components=0.97,svd_solver="full")
pca_mle = PCA(n_components="mle")#实例化自选component
pca_mle = pca_mle.fit(X) #拟合模型
X_mle = pca_mle.transform(X)
# print(X_mle)
print(pca_mle.explained_variance_ratio_.sum())

pca_line = PCA().fit(X)
plt.plot([1,2,3,4], np.cumsum(pca_line.explained_variance_ratio_))
plt.xticks([1,2,3,4])#显示坐标为整数
plt.xlabel("number of components after dimision reduction")
plt.ylabel("explained variance ratio")
```
